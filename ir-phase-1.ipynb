    In this section, implement the clean_data() function. The function works as follows: it takes a text as input and outputs its valid tokens as a list that has undergone lemmatization, stemming, and case folding operations. The input text includes the title of the articles or their abstracts.

```python
# imports
import spacy
import pandas as pd
from nltk.stem import PorterStemmer
import json
import os
import Levenshtein as levenshtein
import math
import heapq
```

```python
# 6 points


def clean_data(text: str):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization, and punctuations

    Parameters
    ----------
    text : str
        The title or abstract of an article

    Returns
    -------
    list
        A list of tokens
    """

    # TODO: tokenize, case_folding, stem, lemmatize, punctuations

    # make text lowercase
    text = text.lower()

    # tokenize
    module = spacy.load("en_core_web_sm")
    txt = module(text)

    # lemmatize
    tokens = [token.lemma_ for token in txt]

    # stem
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]

    # remove punctuations
    tokens = [token for token in tokens if token.isalpha()]

    return tokens


# return ["an", "abstract", "is", "a", "summary", "of", "the", "main", "article"]
clean_data("An abstract is a summary of the main article.")
```

```python
# 2 points


def find_stop_words(all_text: list[str], num_token=30):
    """Detects stop-words

     Parameters
    ----------
    all_text : list of all tokens
        (result of clean_data(text) for all the text)

    Returns
    -------
    Return Value is optional but must print the stop words and number of their occurence
    """

    # TODO: find num_token top most repetitive terms and report them
    counter = {}
    for token in all_text:
        if token not in counter:
            counter[token] = 1
        else:
            counter[token] += 1

    sorted_counter = sorted(counter.items(), key=lambda x: x[1], reverse=True)
    for i in range(num_token):
        print("token: {}, count: {}".format(
            sorted_counter[i][0], sorted_counter[i][1]))

    sorted_counter = sorted_counter[:num_token]
    sorted_counter = [word[0] for word in sorted_counter ]
    return sorted_counter
```

```python
# 12 points

def construct_positional_indexes(corpus):
    """
    Get processed data and insert words in that into a trie and construct postional_index and posting lists after wards.

    Parameters
    ----------
    corpus: str
        processed data 

    Return
    ----------
    docs: 
        list of docs with specificied id, title, abstract.
    """

    # construct positional indexes and posting lists

    # using perfect hash for the term_id
    title_term_id = {}
    abstract_term_id = {}

    # (doc_id, tf)
    title_posting_lists = []
    abstract_posting_lists = []

    # (doc_id, [pos1, pos2, ...])
    title_positional_index = []
    abstract_positional_index = []

    for i in range(len(corpus)):
        title = data.iloc[i, 1]
        abstract = data.iloc[i, 2]
        doc_id = i

        # title 

        if title != "":
            title_tokens = title.split(" ")
            tokens_count = {}
            tokens_position = {}

            # count the number of each token and save the position of each token

            for i in range(len(title_tokens)):
                token = title_tokens[i]
                if token in stop_words:
                    continue
                if token not in tokens_count:
                    tokens_count[token] = 1
                    tokens_position[token] = [i]
                else:
                    tokens_count[token] += 1
                    tokens_position[token].append(i)

            # insert the tokens into the positional indexes and posting lists
            for token in tokens_count:
                if token not in title_term_id:
                    title_term_id[token] = len(title_term_id)
                    title_posting_lists.append([(doc_id, tokens_count[token])])
                    title_positional_index.append(
                        [(doc_id, tokens_position[token])])
                else:
                    title_posting_lists[title_term_id[token]].append(
                        (doc_id, tokens_count[token]))
                    title_positional_index[title_term_id[token]].append(
                        (doc_id, tokens_position[token]))
        
        # abstract 

        # do the same thing for abstract
        if abstract != "":
            abstract_tokens = abstract.split(" ")
            tokens_count = {}
            tokens_position = {}

            # count the number of each token and save the position of each token
            for i in range(len(abstract_tokens)):
                token = abstract_tokens[i]
                if token in stop_words:
                    continue
                if token not in tokens_count:
                    tokens_count[token] = 1
                    tokens_position[token] = [i]
                else:
                    tokens_count[token] += 1
                    tokens_position[token].append(i)
            # insert the tokens into the positional indexes and posting lists
            for token in tokens_count:
                if token not in abstract_term_id:
                    abstract_term_id[token] = len(abstract_term_id)
                    abstract_posting_lists.append(
                        [(doc_id, tokens_count[token])])
                    abstract_positional_index.append(
                        [(doc_id, tokens_position[token])])
                else:
                    abstract_posting_lists[abstract_term_id[token]].append(
                        (doc_id, tokens_count[token])[:100000])
                    abstract_positional_index[abstract_term_id[token]].append(
                        (doc_id, tokens_position[token]))

    return title_term_id, abstract_term_id, title_posting_lists, abstract_posting_lists, title_positional_index, abstract_positional_index
```

    <b>
    query correction
    </b>
    </h1>
</font>
    <br>
    If the input query has a spelling mistake or in other words, some of its words are not in the available dictionary, it is necessary to continue the search by searching for possible words and correcting the query.
    <br>
    <br>
    To do this, first you need to find the bigrams of the word, then with the jaccard criterion, find the twenty words that have the most common bigrams with the word in question. Finally, with the minimum edit distance criterion, find the replacement word for the word in question.
    <br>
    <br>
    There is no need to store and compress the bigram index. Also, you can use ready-made code to calculate the edit distance.

</font>
</div>

```python
def log_tf(tf: int) -> float:

    return 1 + math.log(tf, 2)

def compute_score(query_tf, tf, idf , document_length, avg_doc_length, method):
    l_tf = log_tf(tf)
    l_query_tf = log_tf(query_tf)
    if (method == "ltn-lnn"):
        return l_query_tf * l_tf * idf
    elif (method == "ltn-lnc"):
        return l_query_tf * l_tf * idf / document_length
    elif (method == "bm25"):
        k1 = 1.2
        b = 0.75
        return tf * idf * (k1 + 1) / (tf + k1 * (1 - b + b * document_length / avg_doc_length))
```

```python
from typing import List

def cal_precision(actual , predicted):
    precision = 0.0
    for j in range(len(predicted)):
        if predicted[j] in actual:
            precision += 1
    precision /= len(predicted)
    
    return precision

def calculate_precision(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The precision of the predicted results
    """
    Mean_Precision = 0.0
    for i in range(len(actual)):
        Mean_Precision += cal_precision(actual[i], predicted[i])
    Mean_Precision /= len(actual)
    return Mean_Precision

```

```python
def cal_recall(actual , predicted):
    recall = 0.0
    for i in range(len(actual)):
        if actual[i] in predicted:
            recall += 1
    recall /= len(actual)
    return recall

def calculate_recall(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the recall of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The recall of the predicted results
    """
    Mean_Recall = 0.0
    for i in range(len(actual)):
        Mean_Recall += cal_recall(actual[i], predicted[i])
    Mean_Recall /= len(actual)
    return Mean_Recall
```

```python
def calculate_F1(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the F1 score of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The F1 score of the predicted results    
    """
    f1 = 0.0
    precision = calculate_precision(actual, predicted)
    recall = calculate_recall(actual, predicted)

    f1 = 2 * precision * recall / (precision + recall)
    
    return f1
```
